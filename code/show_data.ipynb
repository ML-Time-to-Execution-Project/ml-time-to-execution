```python
import pandas as pd
```

```python
# lodaing data
message_data = pd.read_csv('../test_data/34200_4200_NFLX_2017-05-01_24900000_57900000_message_50.csv',header = None)
order_data = pd.read_csv('../test_data/34200_4200_NFLX_2017-05-01_24900000_57900000_orderbook_50.csv',header = None)
index_data = pd.read_csv('../test_data/34200_4200_NFLX_2017-05-01_dataset_50.csv',header = 0)
# message_data.head()
# order_data.head()
# index_data.head()
```

```python
# concatenate message file

index_list = index_data.iloc[:,2]
index_data['orderID'] = [message_data.iloc[x,2] for x in index_list]
index_data['size'] = [message_data.iloc[x,3] for x in index_list]
index_data['price'] = [message_data.iloc[x,4] for x in index_list]
index_data['direction'] = [message_data.iloc[x,5] for x in index_list]
index_data.head()
```

```python
# concatenate order book file 
order_size = 5
level_size = 5
 
for i in range(1, order_size+1):
    index_list = [z - i for z in index_list]
    for j in range(level_size):
        for k in range(4):
            index_data[str(i)+str(j)+str(k)] = [order_data.iloc[x,(k+4*j)] for x in index_list]

index_data.head()
```

```python
# split the data into buy and sell

buy_index_data = index_data[index_data['direction'] == 1]
sell_index_data = index_data[index_data['direction'] == -1]
buy_index_data.head()
# sell_index_data.head()
```

```python
buy_index_data.to_csv('../test_data/buy_test_set.csv')
sell_index_data.to_csv('../test_data/sell_test_set.csv')
```

# Order Flow Data

```python
# characteristic extracting
import pandas as pd
```

```python
col_name = []
for i in range(1,51):
    col_name.append('AP' + str(i))
    col_name.append('AS' + str(i))
    col_name.append('BP' + str(i))
    col_name.append('BS' + str(i))
order_data = pd.read_csv('../test_data/34200_4200_NFLX_2017-05-01_24900000_57900000_orderbook_50.csv',names = col_name, header = None)


```

```python
order_diff = order_data.diff()
```

```python
# df.loc[((df.a >0) | (df.b> 0)) & (df.c == '1'), 'd'] = 1
for i in range(1, 51):
   order_data.loc[order_diff['AP'+str(i)] > 0,'aOF'+ str(i)] =  -order_data['AP'+str(i)]
   order_data.loc[order_diff['AP'+str(i)] < 0,'aOF'+ str(i)] =  order_data['AP'+str(i)]
   order_data.loc[order_diff['AP'+str(i)] == 0,'aOF'+ str(i)] = order_diff['AP'+str(i)]
   order_data.loc[order_diff['BP'+str(i)] > 0,'bOF'+ str(i)] =  order_data['BP'+str(i)]
   order_data.loc[order_diff['BP'+str(i)] < 0,'bOF'+ str(i)] =  -order_data['BP'+str(i)]
   order_data.loc[order_diff['BP'+str(i)] == 0,'bOF'+ str(i)] =  order_diff['BP'+str(i)]
order_data.to_csv('../test_data/OF_sample.csv',header=None,index=False)
order_data.head(5)
```
