```python
# script writing
import numpy as np


for lr in np.arange(0.01,0.14,0.04):
    for hidden_size1 in np.arange(50,101,50):
        for hidden_size2 in np.arange(50,101,50):
            for censor_weight in np.arange(0.1,1,0.4):
                for time_weight in np.arange(0.1,1,0.4):
                    for num_epochs in np.arange(80,101,10):
                        path = f'../result_log/parameter_script/lr{lr}_h1{hidden_size1}_h2{hidden_size2}_cw{censor_weight}_tw{time_weight}_epoch{num_epochs}.sh'
                        out = f'#SBATCH -o result_log/MLP_out%j_lr{lr}_h1{hidden_size1}_h2{hidden_size2}_cw{censor_weight}_tw{time_weight}_epoch{num_epochs}.txt\n'
                        err = f'#SBATCH -o result_log/MLP_err%j_lr{lr}_h1{hidden_size1}_h2{hidden_size2}_cw{censor_weight}_tw{time_weight}_epoch{num_epochs}.txt\n'
                        exe = f'python code/version1_fully_connected.py --lr={lr} --hidden_size1={hidden_size1} --hidden_size2={hidden_size2} --censor_weight={censor_weight} --time_weight={time_weight} --num_epochs={num_epochs}\n'
                        print(out)
                        with open(path, "w") as f:
                            f.write('#!/bin/bash\n')
                            f.write('#SBATCH -J MLP\n')
                            f.write(out)
                            f.write(err)
                            f.write('#SBATCH --partition=compute\n')
                            f.write('#SBATCH --qos=low\n')
                            f.write('#SBATCH --get-user-env\n')
                            f.write('#SBATCH --nodes=1\n')
                            f.write('#SBATCH --cpus-per-task=14\n')
                            f.write('#SBATCH --time=36:00:00\n')
                            f.write('## SBATCH --mem=8g\n')
                            f.write('#SBATCH --mail-user=2001212337@stu.pku.edu.cn\n')
                            f.write('#SBATCH --mail-type=all\n')
                            f.write('module load anaconda/3-2021.5\n')
                            f.write('source activate env_deep_learning\n')
                            f.write(exe)
```

```python
# execute script
import os.path
path = f'../result_log/bat_script/MLP_run.bat'
pyname = [name for name in os.listdir('../result_log/parameter_script') if name.endswith('.sh')]

for file in iter(pyname):
    with open(path, "a") as f:
        content = f'sbatch {file}\n'
        f.write(content)


```
